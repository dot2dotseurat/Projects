```python
#Inspecting the By field shows us there are a few extras "by types" (or method for getting on the ballot)
#we'll have to inspect and clean up. 
by_type=ballots.By.unique().tolist()
```

```python
for i in range(len(ballots)):
    if ballots['By'][i]=='SE':
        ballots.loc[i,'By']='se'
    if ballots['By'][i]=='D':
        ballots.loc[i,'By']='S'
    if ballots['By'][i]=='SFUSD':
        ballots.loc[i,'By']='se'
    if ballots['By'][i]=='2':
        ballots.loc[i,'By']='two'
    if ballots['By'][i]=='B':
        ballots.loc[i,'By']='S'
                
```

```python
#Creating a dictionary of each ballot type by year (ie missing years where this method was not used)
ballots_by={}
for item in by_type:
    ballots_by[item]=pd.DataFrame(ballots[ballots.By==item].groupby('Year').count()['Month'])
```

```python
##Creating a dictionary of each ballot type each year
for item in by_type:
    ballots_by[item].columns=[item]
```

```python
#Creating a dataframe for each ballot measure "by type" 
#listing the number of measures each year
dictlist=[]
for key, value in ballots_by.iteritems():
    dictlist.append(value)
```

```python
for i in range(len(dictlist)):
    dictlist[i]=dictlist[i].reset_index()
```

```python
#Yuck but it works to merge all "by type" dataframes together

measures=ballots_by['C'].merge(ballots_by['s'], left_index=True, right_index=True,how='outer')
measures=measures.merge(ballots_by['S'], left_index=True, right_index=True,how='outer')
measures=measures.merge(ballots_by['I'], left_index=True, right_index=True,how='outer')
measures=measures.merge(ballots_by['M'], left_index=True, right_index=True,how='outer')
measures=measures.merge(ballots_by['L'], left_index=True, right_index=True,how='outer')
measures=measures.merge(ballots_by['2'], left_index=True, right_index=True,how='outer')
measures=measures.merge(ballots_by['R'], left_index=True, right_index=True,how='outer')
measures=measures.merge(ballots_by['B'], left_index=True, right_index=True,how='outer')
measures=measures.merge(ballots_by['SE'], left_index=True, right_index=True,how='outer')
measures=measures.merge(ballots_by['D'], left_index=True, right_index=True,how='outer')
measures=measures.merge(ballots_by['se'], left_index=True, right_index=True,how='outer')
measures=measures.merge(ballots_by['SFUSD'], left_index=True, right_index=True,how='outer')
measures=measures.merge(ballots_by['two'], left_index=True, right_index=True,how='outer')
```

```python
#Creating a custom color scheme to show all the methods

import numpy as np
import colorsys

def _get_colors(num_colors):
    colors=[]
    for i in np.arange(0., 360., 360. / num_colors):
        hue = i/360.
        lightness = (45 + np.random.rand() * 10)/100.
        saturation = (90 + np.random.rand() * 10)/100.
        colors.append(colorsys.hls_to_rgb(hue, lightness, saturation))
    return colors
```

```python
#Finally some modeling! Select you X data and create dummy variables

X_data = ballots[['By','Type Measure']]
X_data = pd.get_dummies(X_data)
```

```python
#Select your y data
y_data = ballots['Pass_Num']
```

```python
#Split your data into test and train sets. Fit to your train set and score on your test set.

from sklearn.cross_validation import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_data, y_data, random_state=12, test_size=0.3)
from sklearn.svm import LinearSVC
clf = LinearSVC()
clf.fit(X_train, y_train)
clf.score(X_test, y_test)
```

```python
#Print your confusion matrix showing true and false postives and negatives. 
#Bottom right true positives and top left true negatives.
from sklearn.metrics import confusion_matrix, classification_report
y_pred = clf.predict(X_test)
confusion_matrix(y_test, y_pred)
```

```python
#Ordered feature importance
pd.DataFrame(zip(X_data.columns, np.transpose(clf.coef_))).sort(1, ascending=False)
```
